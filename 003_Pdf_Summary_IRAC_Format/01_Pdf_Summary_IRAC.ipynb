````python
# Import Packages
import openai
import os
import PyPDF2

# Assigning API key to the variable openai.api_key
openai.api_key = os.getenv('OPENAI_API_kEY')
````

````python
# defining helper_function to get the response from ChatGPT
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0,
    )
    return response.choices[0].message["content"]
````

````python
# Defining the readPdfFile function to read the pdf file and output as text
def readPdfFile(filePath):
    with open(filePath, 'rb') as file:
        page_reader = PyPDF2.PdfReader(file)
        num_pages = len(page_reader.pages)
        ans = ""
        for page_num in range(num_pages):
            page = page_reader.pages[page_num]
            text = page.extract_text()
            ans += text
    return ans
````

````python
# To read the pdf file and convert to the text format
text = readPdfFile(".\Bank of Maharashtra vs Directorate of Revenue Intelligence (Bombay HC).pdf")
print(text)
````

````python
prompt = f"""Your task is to generate IRAC format of the given text \
        delimited by triple backticks.\
        Case Summary:
             ```{text}```
            """
response = get_completion(prompt)
print(response) 
````

````python
prompt = f"""Perform the following actions:\
        1. Write down the issues of the given text.
        2. List out the rules of the given text with bullet points.
        3. Analyze the given text.
        4. Write the conclusion of the given text.
        
        Use the following format:
        Issues : <<<Issue>>>
        Rules : <<<Rules>>>
        Analysis : <<<Analysis>>>
        Conclusion : <<<Conclusion>>>
        
        Separate it with line breaks.
        
        ```{text}```
            """
response = get_completion(prompt)
print(response)           
````
